```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
T0 = 80  # Initial temperature of the object in °C
T_inf = 20  # Ambient temperature in °C
rho = 1000  # Density of the object in kg/m^3
C = 2000  # Heat capacity in J/K
A = 0.5  # Surface area in m^2
V = 0.25  # Volume in m^3
time = np.arange(0, 121, 1)  # Time from 0 to 120 seconds
```

```python
# Heat transfer coefficients
h_values = [5, 10, 15]
temperature_results = {}

# Calculate the temperature for each h value
for h in h_values:
    T = T_inf + (T0 - T_inf) * np.exp(-h * A * time / (rho * V * C))
    temperature_results[h] = T
```

#Plot Graph

```python

# Plotting the results
plt.figure(figsize=(10, 6))
for h in h_values:
    plt.plot(time, temperature_results[h], marker='o', linestyle='-', label=f'h = {h} W/m²K')

# Customize the plot
plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Cooling of a Cylindrical Object using Lumped Capacitance Method')
plt.legend()
plt.grid(True)
plt.show()
```
